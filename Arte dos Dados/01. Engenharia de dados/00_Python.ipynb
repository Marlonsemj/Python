# Regras gerais para nomes de variáveis no Python
* As variáveis no Python devem começar com letras (A-Z/a-z) ou underscore(_)
* As variáveis Python não devem conter símbolos especiais como, por exemplo, !, @, #, $, %, .
* Palavras reservadas do Python não podem ser usadas como nomes de variáveis;
* Observação: Python é case-sensitive. Portanto, a variável "nome" e "Nome" são variáveis distintas.

### Exemplo 4: variáveis e estrutura usando PEP8
Que tal a documentação do código abaixo?

```python
x = 'Pedro Alvares Cabral'   # x é a variável que definimos ser "Pedro Alvares Cabral"
a, b, c = x.split()          # O comando split irá dividir a partir do momento que encontrar espaços para cada variável
print(a)                     # O comando print irá imprimir a variável a, que aqui é Pedro
print(b)                     # O comando print irá imprimir a variável b, que aqui é Alvares
print(c)                     # O comando print irá imprimir a variável c, que aqui é Cabral
```

*Que tal dessa forma...*

```python
s_descobridor_Brasil = 'Pedro Alvares Cabral'
s_primeiro_nome, s_segundo_nome, s_terceiro_nome = s_descobridor_Brasil.split()
print(s_primeiro_nome)
print(s_segundo_nome)
print(s_terceiro_nome)
# essa forma é, provavelmente, a mais fácil de entender
```

> As classes em Python são um dos principais conceitos da programação orientada a objetos (OOP). Elas servem como um molde ou modelo para criar objetos que podem armazenar dados e comportamentos. Em termos simples, as classes permitem organizar e estruturar o código de forma mais eficiente, reutilizável e modular.
>
> ### **O que é uma Classe?**
>
> Uma classe define um tipo de objeto, especificando suas propriedades (chamadas de atributos) e suas ações (chamadas de métodos). Quando você cria uma classe, está criando uma "forma" ou "esqueleto" que pode ser usado para criar instâncias (objetos) daquela classe.

```python
# Definição da classe
class Carro:
    # Método especial __init__ para inicializar os atributos da classe
    def __init__(self, marca, modelo, ano):
        self.marca = marca  # Atributo marca
        self.modelo = modelo  # Atributo modelo
        self.ano = ano  # Atributo ano

    # Método para exibir informações sobre o carro
    def exibir_detalhes(self):
        print(f"Carro: {self.marca} {self.modelo}, Ano: {self.ano}")

# Criando uma instância (objeto) da classe Carro
meu_carro = Carro("Toyota", "Corolla", 2020)

# Chamando um método da classe
meu_carro.exibir_detalhes()
```
